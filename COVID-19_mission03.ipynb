# COVID-19 ANALYSIS (INDIA)

To Represent the Covid Cases State-Wise(India) of 04/2020 - 06/2020 through Exploratory Data Analysis.

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Data Collection

```python
cases=pd.read_csv('C:/Users/Ujjwal/Downloads/Widhya/COVID-19_mission03.csv')
```

## Data Cleaning & Filling

```python
cases.fillna(value= cases.mean(), inplace= True)
```

```python
cases=cases.drop_duplicates()
```

```python
f,ax = plt.subplots(figsize = (14,10)) 
sns.heatmap(cases.isnull(),annot=True,ax=ax, cmap= 'coolwarm',linewidths=4, linecolor='white')
```

## Exploratory Data Analysis

```python
cases.drop(["S. No."],axis=1,inplace=True)
```

```python
cases.rename(columns={"Name of State / UT":"State", "Cured/Discharged/Migrated":"Discharged","Active Cases":"Active"},inplace=True)
```

```python
cases.corr()
```

```python
f,ax = plt.subplots(figsize = (7,7)) 
sns.heatmap(cases.corr(),annot=True,ax=ax,linewidths=4, linecolor='white')
```

## State-Wise Confirmed Cases

```python
plt.figure(figsize=(15,10))
plt.bar(cases["State"],cases["Total Confirmed cases"])

plt.title("Total Confirmed Cases",fontsize=16)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Patients",fontsize=15)

plt.xticks(rotation=90)
plt.yticks()
plt.show()
```

## State-Wise Discharged Patients

```python
plt.figure(figsize=(15,10))
plt.bar(cases["State"],cases["Discharged"])

plt.title("Total Discharged Patients",fontsize=16)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Patients",fontsize=15)
plt.xticks(rotation=90)

plt.show()
```

## State-Wise Active Cases

```python
plt.figure(figsize=(15,10))
plt.bar(cases["State"],cases["Active"])

plt.title("Total Active Cases By State",fontsize=16)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Cases",fontsize=15)

plt.xticks(rotation=90)
plt.yticks()
plt.show()
```

## State-Wise Death Cases

```python
plt.figure(figsize=(15,10))
plt.bar(cases["State"],cases["Deaths"])

plt.title("Total Death Cases By State",fontsize=16)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Cases",fontsize=15)

plt.xticks(rotation=90)
plt.yticks()
plt.show()
```

##  Risk Level

Here, we will check the risk level of Covid-19 Virus.  

#### Discharge and Death Cases

```python
f, ax=plt.subplots(figsize=(10,7))
cases['Deaths'].plot(color='blue', label = 'Deaths', linewidth=2, grid = True, linestyle = '-')
cases['Discharged'].plot(color = 'red', label = 'Discharge', linewidth=1.2, grid = True, linestyle = '--')

plt.legend(loc=0)
plt.xlabel('Deaths')
plt.ylabel('Discharge')
plt.title('Line Plot', size=16)
```

Through this, we can see discharges are more than deaths.

#### Death Rate

Here we will calculate Death Rate by Covid-19.

```python
Death_Rate = (cases['Deaths']/cases['Total Confirmed cases'])*100
```

```python
plt.figure(figsize=(15,10))
plt.plot(cases["State"], Death_Rate)

plt.xlabel('State / Union terirorry', size=15,color='green')
plt.ylabel('Death Rate(%)', size=15,color='red')
plt.title('Death Rate', size = 18)


plt.xticks(rotation=90)
plt.yticks()
plt.grid(True)
```

We can clearly see that the Death rate is less than 3 percent. Which is good sign, if we follow government rules and regulatories seriously, we will be able to defeat Covid-19 with ease.

## India Covid-19 Overview

```python
most_effected_states = cases.groupby('State')[['Active','Deaths','Discharged','Total Confirmed cases']].sum().sort_values(by='Total Confirmed cases',ascending=False)
most_effected_states.head(10)
```

These states have the most confirmed Covid-19 Cases. Indian Government should strictly practice covid rules and regulations in these states.

```python
Deaths=cases['Deaths'].sum()
Discharged=cases['Discharged'].sum()
Active=cases['Active'].sum()
Total_Cases=cases["Total Confirmed cases"].sum()
```

```python
Report = [Active,Deaths,Discharged,Total_Cases]
labels = ['Active','Deaths','Discharged','Total_Cases']
colors = ['orange', 'cyan', 'green', 'grey'] 

f,ax=plt.subplots(figsize=(12,12))
plt.title('INDIA COVID-19 OVERVIEW', size=30, fontname="Times New Roman", fontweight="bold")
plt.pie( Report, autopct='%1.1f%%', labels=labels, colors= colors)

plt.legend(loc=(1.05,0.5), prop={'size': 15})
```
